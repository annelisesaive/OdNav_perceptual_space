```python
def detect_outliers_pearson(column_a, column_b, epsilon):
    '''This function detects outliers given two arrays, calculates the coefficient of correlation of Pearson
    and draws a regression plot if the correlation is significative, i.e. the p-value is <0.1'''
    df = pd.DataFrame({"feature" : column_a, "number" : column_b})
    z_scores = stats.zscore(df)
    abs_z_scores = np.abs(z_scores)
    filtered_entries = (abs_z_scores < epsilon).all(axis=1)
    df_wo_outliers = df[filtered_entries]
    pearson = pearsonr(df_wo_outliers["feature"], df_wo_outliers["number"])
    if pearson[1] < 0.1:
        print("Significant")
        print("Pearson coefficient = " + str(pearson[0]) + ", p-value = " + str(pearson[1]))
        p = sns.regplot(x = df_wo_outliers["feature"], y = df_wo_outliers["number"]) 
    return(pearson)
```

```python
def detect_outliers_kendall(column_a, column_b, epsilon):
    '''This function detects outliers given two arrays, calculates the coefficient of correlation of Kendall
    and draws a regression plot if the correlation is significative, i.e. the p-value is <0.1'''
    df = pd.DataFrame({"feature" : column_a, "number" : column_b})
    z_scores = stats.zscore(df)
    abs_z_scores = np.abs(z_scores)
    filtered_entries = (abs_z_scores < epsilon).all(axis=1)
    df_wo_outliers = df[filtered_entries]
    kendall = kendalltau(df_wo_outliers["feature"], df_wo_outliers["number"])
    if kendall[1] < 0.1:
        print("Significant")
        print("Kendall coefficient = " + str(kendall[0]) + ", p-value = " + str(kendall[1]))
        p = sns.regplot(x = df_wo_outliers["feature"], y = df_wo_outliers["number"]) 
    return(kendall)
```

```python
def get_filepaths(directory):
    """
    This function will generate the file names in a directory 
    tree by walking the tree either top-down or bottom-up. For each 
    directory in the tree rooted at directory top (including top itself), 
    it yields a 3-tuple (dirpath, dirnames, filenames).
    """
    file_paths = []
    for root, directories, files in os.walk(directory):
        for filename in files:
            filepath = os.path.join(root, filename)
            file_paths.append(filepath)
    return(file_paths)
```

```python
def calculateDistance(x1,y1,x2,y2):
    '''This function calculates the distance between two points given their coordinates'''
    dist = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return(dist)
```

```python
df_scores = pd.DataFrame({"subject" : np.concatenate([kept_subjects, ["S08-BSC", "S14-BSC"]])})

for i in range(51):
    df_scores.at[i,"n_hit"] = features_rappel[i][features_rappel[i]['score_recognition'] == 'hit']['trial'].count()
    df_scores.at[i,"n_fa"] = features_rappel[i][features_rappel[i]['score_recognition'] == 'fa']['trial'].count()
    df_scores.at[i,"n_cr"] = features_rappel[i][features_rappel[i]['score_recognition'] == 'cr']['trial'].count()
    df_scores.at[i,"n_miss"] = features_rappel[i][features_rappel[i]['score_recognition'] == 'miss']['trial'].count()
    df_scores.at[i,"n_w"] = features_rappel[i][features_rappel[i]['episodic_score'] == 'w']['trial'].count()
    df_scores.at[i,"n_ww"] = features_rappel[i][features_rappel[i]['episodic_score'] == 'ww']['trial'].count()
    df_scores.at[i,"n_www"] = features_rappel[i][features_rappel[i]['episodic_score'] == 'www']['trial'].count()
    df_scores.at[i,"HR"] = (df_scores.at[i,"n_hit"] + 0.5) / (df_scores.at[i,"n_hit"] + df_scores.at[i,"n_miss"] + 1)
    df_scores.at[i,"FR"] = (df_scores.at[i,"n_fa"] + 0.5) / (df_scores.at[i,"n_cr"] + df_scores.at[i,"n_fa"] + 1)
    df_scores.at[i,"d_prime"] = np.log((df_scores.at[i,"HR"] * (1 - df_scores.at[i,"FR"])) / 
                                       (df_scores.at[i,"FR"] * (1 - df_scores.at[i,"HR"])))
    df_scores.at[i,"W_score"] = df_scores.at[i,"n_w"] / df_scores.at[i,"n_hit"]
    df_scores.at[i,"WW_score"] = df_scores.at[i,"n_ww"] / df_scores.at[i,"n_hit"]
    df_scores.at[i,"WWW_score"] = df_scores.at[i,"n_www"] / df_scores.at[i,"n_hit"]
```

```python
data_distance = pd.DataFrame({"subject" : all_subjects, 
                              "hit_o" : np.repeat(np.NAN, 51), "hit_m" : np.repeat(np.NAN, 51), 
                              "hit_f" : np.repeat(np.NAN, 51), "fa_o" : np.repeat(np.NAN, 51),
                              "fa_m" : np.repeat(np.NAN, 51), "fa_f" : np.repeat(np.NAN, 51),
                              "w_o" : np.repeat(np.NAN, 51), "w_m" : np.repeat(np.NAN, 51),
                              "w_f" : np.repeat(np.NAN, 51), "ww_o" : np.repeat(np.NAN, 51), 
                              "ww_m" : np.repeat(np.NAN, 51), "ww_f" : np.repeat(np.NAN, 51), 
                              "www_o" : np.repeat(np.NAN, 51), "www_m" : np.repeat(np.NAN, 51), 
                              "www_f" : np.repeat(np.NAN, 51)})
```

```python
for i in range(51):
    mean = features_rappel[i].groupby(['score_recognition','stimulus_type']).mean()
    data_distance.at[i,"hit_o"] = mean.loc[("hit","odor")]["distance"]
    data_distance.at[i,"hit_f"] = mean.loc[("hit","face")]["distance"]
    data_distance.at[i,"hit_m"] = mean.loc[("hit","music")]["distance"]
    if ("fa","odor") in mean.index.tolist():
        data_distance.at[i,"fa_o"] = mean.loc[("fa","odor")]["distance"]
    if ("fa","face") in mean.index.tolist():
        data_distance.at[i,"fa_f"] = mean.loc[("fa","face")]["distance"]
    if ("fa","music") in mean.index.tolist():
        data_distance.at[i,"fa_m"] = mean.loc[("fa","music")]["distance"]
```

```python
for i in range(51):
    feature = features_rappel[i]
    mean = feature.groupby(['episodic_score','stimulus_type']).mean()
    if ("w","odor") in mean.index.tolist():
        data_distance.at[i,"w_o"] = mean.loc[("w","odor")]["distance"]
    if ("w","face") in mean.index.tolist():
        data_distance.at[i,"w_f"] = mean.loc[("w","face")]["distance"]
    if ("w","music") in mean.index.tolist():
        data_distance.at[i,"w_m"] = mean.loc[("w","music")]["distance"]

    if ("ww","odor") in mean.index.tolist():
        data_distance.at[i,"ww_o"] = mean.loc[("ww","odor")]["distance"]
    if ("ww","face") in mean.index.tolist():
        data_distance.at[i,"ww_f"] = mean.loc[("ww","face")]["distance"]
    if ("ww","music") in mean.index.tolist():
        data_distance.at[i,"ww_m"] = mean.loc[("ww","music")]["distance"]
        
    if ("www","odor") in mean.index.tolist():
        data_distance.at[i,"www_o"] = mean.loc[("www","odor")]["distance"]
    if ("www","face") in mean.index.tolist():
        data_distance.at[i,"www_f"] = mean.loc[("www","face")]["distance"]
    if ("www","music") in mean.index.tolist():
        data_distance.at[i,"www_m"] = mean.loc[("www","music")]["distance"]
```

- We'll get rid of subjects with missing values because the friedman test requires replicated blocks:

```python
sub_to_discard = np.array([])
for i in np.where(np.isnan(anova_distance_recognition.distance.values))[0]:
    subject = anova_distance_recognition.subject.values[i]
    sub_to_discard = np.append(sub_to_discard, subject)
```

```python
rows_to_discard = np.array([])
for sub in sub_to_discard:
    ind = anova_distance_recognition[anova_distance_recognition.subject == sub].index.tolist()
    rows_to_discard = np.append(rows_to_discard, ind)
```

```python
anova_distance_recognition = anova_distance_recognition.drop(rows_to_discard)
anova_distance_recognition = anova_distance_recognition.reset_index(drop=True)
```

```python
anova_distance_recognition.to_csv("Created datasets/anova_distance_recognition.csv")
```

## <font color=blue>Calculating the shortest path to each box</font><a id="part9sec1"></a>

```python
def intercept(c1, c2, r, a, b):
    '''
    This function returns the coordinates of the two intersections between a circle and a line, given the 
    coordinates of the center (c1, c2) and radius (r) of the former and the slope (a) and y-intercept (b) of the
    latter.
    '''
    delta = 4*(a*b-a*c2-c1)**2-4*(a**2+1)*((b-c2)**2+c1**2-r**2)
    x1 = (2*(c1+a*c2-a*b) - np.sqrt(delta))/(2*(a**2+1))
    y1 = a*x1+b
    x2 = (2*(c1+a*c2-a*b) + np.sqrt(delta))/(2*(a**2+1))
    y2 = a*x2+b
    return([[x1, y1], [x2, y2]])

def line(x1,y1,x2,y2):
    '''
    This function returns the slope and the y-intercept of a line given the coordinates of two of its points
    (x1, y1) and (x2, y2)
    '''
    slope = (y2-y1)/(x2-x1)
    y_intercept = y1 - slope * x1
    return([slope,y_intercept])

def orth(x1,y1,x2,y2,x3,y3):
    '''
    This function returns the slope and the y-intercept of a line perpendicular to another line (linking two 
    points (x2, y2) and (x3,y3)) and passing through a point (x1, y1)
    '''
    slope = line(x2,y2,x3,y3)[0]
    slope_orth = - (1/slope)
    y_intercept_orth = y1 - slope_orth * x1
    return([slope_orth, y_intercept_orth])

def par(x1,y1,slope):
    '''
    This function returns the slope and the y-intercept of a line parallel to another one with a known slope
    and passing through a point (x1, y1)
    '''
    slope_par = slope
    y_intercept_par = y1 - slope_par * x1
    return([slope_par, y_intercept_par])

def abline(slope, intercept, x1, x2):
    '''
    This function draws a segment given its slope and its y-intercept limited by the abscissas x1 and x2
    '''
    axes = plt.gca()
    x_vals = np.linspace(x1,x2)
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
    
def intersection(a1,b1,a2,b2):
    '''
    This function returns the coordinates of the intersection of two lines giveen their slopes (a1 and a2) and
    their y-intercepts (b1 and b2)
    '''
    x = (b2-b1)/(a1-a2)
    y = a1 * x + b1
    return([x,y])

def intercercle(x0, y0, r0, x1, y1, r1):
    '''
    This function returns the coordinates of the intersections of two circles given their centres (x0, y0) and 
    (x1, y1) and their radii r0 and r1
    '''
    if y0 == y1: 
        x = (r1**2 - r0**2 + x0**2 - x1**2)/(2*(x0-x1))
        y_a = y0 - np.sqrt(r0**2 - (x-x0)**2)
        y_b = y0 + np.sqrt(r0**2 - (x-x0)**2)
        return([x, y_a], [x, y_b])
    else:
        n = (r1**2 - r0**2 - x1**2 + x0**2 - y1**2 + y0**2)/(2*(y0 - y1))
        a = ((x0 - x1)/(y0 - y1))**2 + 1
        b = 2 * y0 * ((x0 - x1)/(y0 - y1)) - 2 * n * ((x0 - x1)/(y0 - y1)) - 2 * x0
        c = x0**2 + y0**2 + n**2 - r0**2 - 2 * y0 * n
        delta = b**2 - 4 * a * c
        x_a = (-b - np.sqrt(delta))/(2*a)  
        y_a = n - x_a * ((x0 - x1)/(y0 - y1))
        x_b = (-b + np.sqrt(delta))/(2*a)
        y_b = n - x_b * ((x0 - x1)/(y0 - y1))
        return([x_a, y_a], [x_b, y_b])
```

- Below are some projection functions that we'll need to calculate the deviation:

```python
def proj_seg(x1,y1,x2,y2,x3,y3):
    '''
    This function returns the coordinates of the projection of the point (x1, y1) onto the segment bounded by
    the points (x2, y2) and (x3, y3). It also returns the distance between the point and the segment support 
    and a boolean indicating whether the projected point belongs to the segment
    '''
    if y2 == y3:
        p = [x1, y2]
        boolean = False
        d = calculateDistance(p[0], p[1], x1, y1)
        if (p[0] >= min(x2,x3) and p[0] <= max(x2,x3)):
            boolean = True
        return([p, d, boolean])
    else:
        l = line(x2,y2,x3,y3)
        o = orth(x1,y1,x2,y2,x3,y3)
        p = intersection(l[0], l[1], o[0], o[1])
        d = calculateDistance(p[0], p[1], x1, y1)
        boolean = False
        if (p[0] >= min(x2,x3) and p[0] <= max(x2,x3) and p[1] >= min(y2,y3) and p[1] <= max(y2,y3) 
            and p[1] == l[0]*p[0]+l[1]):
            boolean = True
        return([p, d, boolean])

def proj_demi_gauche(x1,y1,x2,y2,x3,y3):
    '''
    This function returns the coordinates of the projection of the point (x1, y1) onto the left half line 
    (x2, y2)  and (x3, y3). It also returns the distance between the point and the support of the half-line and 
    a boolean indicating whether the projection belongs to this half-line
    '''
    l = line(x2,y2,x3,y3)
    o = orth(x1,y1,x2,y2,x3,y3)
    p = intersection(l[0], l[1], o[0], o[1])
    d = calculateDistance(p[0], p[1], x1, y1)
    boolean = False
    if (p[0] <= max(x2,x3) and p[1] == l[0]*p[0]+l[1]):
        boolean = True
    return([p, d, boolean])

def proj_demi_droite(x1,y1,x2,y2,x3,y3):
    '''
    This function returns the coordinates of the projection of the point (x1, y1) onto the right half line
    (x2, y2) and (x3, y3). It also returns the distance between the point and the support of the half-line and
    a boolean indicating whether the projection belongs to this half-line
    '''
    l = line(x2,y2,x3,y3)
    o = orth(x1,y1,x2,y2,x3,y3)
    p = intersection(l[0], l[1], o[0], o[1])
    d = calculateDistance(p[0], p[1], x1, y1)
    boolean = False
    if (p[0] >= min(x2,x3) and p[1] == l[0]*p[0]+l[1]):
        boolean = True
    return([p, d, boolean])

def proj_cerc_gauche(x1, y1, x2, y2, x3, y3, c1, c2, r):
    if x1 == c1:
        p = [c1, c2+r]
        d = calculateDistance(x1,y1,c1,c2)
        boolean = False
        if (p[0] >= min(x2,x3) and p[0] <= max(x2,x3) and p[1] <= max(y2,y3) 
            and (p[0]+c1)**2+(p[1]-c2)**2-r**2 <= 0.001):
            boolean = True
        return([p, np.abs(d-r), boolean])
    else:
        l = line(x1,y1,c1,c2)
        d = calculateDistance(x1,y1,c1,c2)
        p = intercept(c1, c2, r, l[0], l[1])[0]
        boolean = False
        if (p[0] >= min(x2,x3) and p[0] <= max(x2,x3) and p[1] <= max(y2,y3)
            and (p[0]-c1)**2+(p[1]-c2)**2-r**2 <= 0.001):
            boolean = True
        return([p, np.abs(d-r), boolean]) 
    
def proj_cerc_gauche_bis(x1, y1, x2, y2, x3, y3, c1, c2, r):
    if x1 == c1:
        p = [c1, c2+r]
        d = calculateDistance(x1,y1,c1,c2)
        boolean = False
        if (p[0] >= min(x2,x3) and p[0] <= max(x2,x3) and p[1] >= min(y2,y3) 
            and (p[0]+c1)**2+(p[1]-c2)**2-r**2 <= 0.001):
            boolean = True
        return([p, np.abs(d-r), boolean])
    else:
        l = line(x1,y1,c1,c2)
        d = calculateDistance(x1,y1,c1,c2)
        p = intercept(c1, c2, r, l[0], l[1])[0]
        boolean = False
        if (p[0] >= min(x2,x3) and p[0] <= max(x2,x3) and p[1] >= min(y2,y3)
            and (p[0]-c1)**2+(p[1]-c2)**2-r**2 <= 0.001):
            boolean = True
        return([p, np.abs(d-r), boolean]) 
    
def proj_cerc_droite(x1, y1, x2, y2, x3, y3, c1, c2, r):
    if x1 == c1:
        p = [c1, c2-r]
        d = calculateDistance(x1,y1,c1,c2)
        boolean = False
        if (p[0] >= min(x2,x3) and p[0] <= max(x2,x3) and p[1] <= max(y2,y3) 
            and (p[0]-c1)**2+(p[1]-c2)**2-r**2 <= 0.001):
            boolean = True
        return([p, np.abs(d-r), boolean])
    else:
        l = line(x1,y1,c1,c2)
        d = calculateDistance(x1,y1,c1,c2)
        p = intercept(c1, c2, r, l[0], l[1])[1]
        boolean = False
        if (p[0] >= min(x2,x3) and p[0] <= max(x2,x3) and p[1] <= max(y2,y3)
            and (p[0]-c1)**2+(p[1]-c2)**2-r**2 <= 0.001):
            boolean = True
        return([p, np.abs(d-r), boolean])
    
def proj_cerc_droite_bis(x1, y1, x2, y2, x3, y3, c1, c2, r):
    if x1 == c1:
        p = [c1, c2-r]
        d = calculateDistance(x1,y1,c1,c2)
        boolean = False
        if (p[0] >= min(x2,x3) and p[0] <= max(x2,x3) and p[1] >= min(y2,y3) 
            and (p[0]-c1)**2+(p[1]-c2)**2-r**2 <= 0.001):
            boolean = True
        return([p, np.abs(d-r), boolean])
    else:
        l = line(x1,y1,c1,c2)
        d = calculateDistance(x1,y1,c1,c2)
        p = intercept(c1, c2, r, l[0], l[1])[1]
        boolean = False
        if (p[0] >= min(x2,x3) and p[0] <= max(x2,x3) and p[1] >= min(y2,y3)
            and (p[0]-c1)**2+(p[1]-c2)**2-r**2 <= 0.001):
            boolean = True
        return([p, np.abs(d-r), boolean]) 
```

```python
for i in range(51):
    data_patterns[i]["patterns_3"] = np.empty(3, dtype = object)
```

```python
for i in range(51):
    for j in range(3):
        path = data_patterns[i].path[j]
        liste_patterns = []
        for k in range(len(path)-2):
            liste_patterns.append([path[k], path[k+1], path[k+2]])
        data_patterns[i].at[j,"patterns_3"] = liste_patterns
```

```python
for i in range(51):
    data_patterns[i]["all_different"] = np.empty(3, dtype = int)
    data_patterns[i]["two_similar_cons"] = np.empty(3, dtype = int)
    data_patterns[i]["two_similar_sep"] = np.empty(3, dtype = int)
    data_patterns[i]["all_similar"] = np.empty(3, dtype = int)
    data_patterns[i]["prop_different"] = np.empty(3, dtype = float)
    data_patterns[i]["prop_two_similar_cons"] = np.empty(3, dtype = float)
    data_patterns[i]["prop_two_similar_sep"] = np.empty(3, dtype = float)
    data_patterns[i]["prop_similar"] = np.empty(3, dtype = float)
```

```python
for i in range(51):
    df = data_patterns[i]
    for j in range(3):
        patterns = df.patterns_3[j]
        cpt_all_diff = 0
        cpt_all_sim = 0
        cpt_two_sim_consecutive = 0
        cpt_two_sim_separated = 0
        for elt in patterns: 
            if (elt[0] != elt[1]) and (elt[1] != elt[2]) and (elt[0] != elt[2]):
                cpt_all_diff += 1
            elif elt[0] == elt[1] == elt[2]:
                cpt_all_sim += 1
            elif (elt[0] != elt[1]) and (elt[1] != elt[2]) and (elt[0] == elt[2]):
                cpt_two_sim_separated += 1
            else:
                cpt_two_sim_consecutive += 1 
        df.all_different[j] = cpt_all_diff
        df.two_similar_cons[j] = cpt_two_sim_consecutive
        df.two_similar_sep[j] = cpt_two_sim_separated
        df.all_similar[j] = cpt_all_sim
```

```python
for i in range(51):
    df = data_patterns[i]
    for j in range(3):
        df.prop_different[j] = df.all_different[j] / len(df.patterns_3[j])
        df.prop_two_similar_cons[j] = df.two_similar_cons[j] / len(df.patterns_3[j])
        df.prop_two_similar_sep[j] = df.two_similar_sep[j] / len(df.patterns_3[j])
        df.prop_similar[j] = df.all_similar[j] / len(df.patterns_3[j])
```

```python
prop_diff = np.empty(51)
prop_sim = np.empty(51)
prop_two_sim_cons = np.empty(51)
prop_two_sim_sep = np.empty(51)
for i in range(51):
    prop_diff[i] = data_patterns[i].all_different.sum() / sum([len(data_patterns[i].patterns_3[j])
                                                               for j in range(3)])
    prop_sim[i] = data_patterns[i].all_similar.sum() / sum([len(data_patterns[i].patterns_3[j])
                                                            for j in range(3)])
    prop_two_sim_cons[i] = (data_patterns[i].two_similar_cons.sum() / 
                            sum([len(data_patterns[i].patterns_3[j]) for j in range(3)]))
    prop_two_sim_sep[i] = (data_patterns[i].two_similar_sep.sum() / 
                           sum([len(data_patterns[i].patterns_3[j]) for j in range(3)]))
```

```python
def unique(list1):
    '''
    This function returns the unique elements in a list even if these elements are not the same type. The
    function numpy.unique isn't enough in this case, which is why we define our own.
    '''
    liste = []
    list_set = set(list1)
    unique_list = (list(list_set))
    for x in unique_list:
        liste.append(x)
    return(np.array(liste))
```

## <font color=blue>Distance matrices</font><a id="part15sec7"></a>

```python
from scipy.spatial import distance_matrix
```

```python
real_distance_matrix = np.empty(49, dtype = object)
perc_distance_matrix = np.empty(49, dtype = object)
for i in range(49):
    real_distance_matrix[i] = np.empty(3, dtype = object)
    perc_distance_matrix[i] = np.empty(3, dtype = object)
    for j in range(3):
        df_pca = data_physical[i][j]
        df_iter_pca = data_pca[i][j]
        a = pd.DataFrame(distance_matrix(df_pca[["box_x", "box_y"]].values, df_pca[["box_x", "box_y"]].values), 
                         index = df_pca.stimulus_code, columns = df_pca.stimulus_code)
        b = pd.DataFrame(distance_matrix(df_iter_pca[["dim_1", "dim_2"]].values, 
                                         df_iter_pca[["dim_1", "dim_2"]].values), 
                         index = df_iter_pca.stimulus_code, columns = df_iter_pca.stimulus_code)
        real_distance_matrix[i][j] = a
        perc_distance_matrix[i][j] = b
```

```python
correlation_pca = np.empty(49, dtype = object)
for i in range(49):
    correlation_pca[i] = np.empty(3)
    for j in range(3):
        a = real_distance_matrix[i][j].values
        b = perc_distance_matrix[i][j].values
        correlation_pca[i][j] = pearsonr(a[np.triu_indices(9)], b[np.triu_indices(9)])[0]
```
